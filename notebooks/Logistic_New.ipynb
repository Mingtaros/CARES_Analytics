```python
cares_data_path = "C:/Users/aksha/OneDrive/Documents/mitb_sem2.5/Health_Analytics/Health_Leo/CARES_encoded_ignorecols.csv"
data = pd.read_csv(cares_data_path)
data.head()
```

```python
import pandas as pd
import statsmodels.api as sm

# Define the dependent and independent variables
dependent_var = 'ICUAdmgt24h'
independent_vars = ['AGEcategory', 'GENDER', 'Anemia category', 'Perioperativetransfusionwithin30days', 
                    'AnaestypeCategory', 'SurgRiskCategory', 'PriorityCategory', 'GradeofKidneyCategory', 
                    'RDW15.7', 'ASAcategorybinned', 'Chinese', 'Indian', 'Malay','CVARCRICategory','IHDRCRICategory',
                   'CHFRCRICategory','DMinsulinRCRICategory','CreatinineRCRICategory']

# Prepare the data
X = data[independent_vars]
y = data[dependent_var]

# Add a constant to the independent variables (intercept)
X = sm.add_constant(X)

# Fit the logistic regression model
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Display the summary of the logistic regression
print(result.summary())
```

<b>The significant positive predictors increase the likelihood of ICU admission, meaning these factors are associated with a higher risk of needing intensive care.<br>Significant Positive Predictors: AGEcategory, GENDER, Perioperativetransfusionwithin30days, SurgRiskCategory, PriorityCategory, GradeofKidneyCategory, ASAcategorybinned, and CHFRCRICategory.<br><br>
Significant negative predictors decrease the likelihood, indicating that these factors are associated with a lower risk of ICU admission.<br>Significant Negative Predictors: AnaestypeCategory, DMinsulinRCRICategory, and CreatinineRCRICategory.<br><br>
Non-significant predictors did not have a statistically meaningful impact on ICU admission in this model.<br>Non-Significant Predictors: Anemia category, RDW15.7, CVARCRICategory, IHDRCRICategory, Indian, and Malay

```python
import pandas as pd
import statsmodels.api as sm

# Define the dependent and independent variables
dependent_var = 'Mortality'
independent_vars = ['AGEcategory', 'GENDER', 'Anemia category', 'Perioperativetransfusionwithin30days', 
                    'AnaestypeCategory', 'SurgRiskCategory', 'PriorityCategory', 'GradeofKidneyCategory', 
                    'RDW15.7', 'ASAcategorybinned', 'Chinese', 'Indian', 'Malay','CVARCRICategory','IHDRCRICategory',
                   'CHFRCRICategory','DMinsulinRCRICategory','CreatinineRCRICategory']

# Prepare the data
X = data[independent_vars]
y = data[dependent_var]

# Add a constant to the independent variables (intercept)
X = sm.add_constant(X)

# Fit the logistic regression model
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Display the summary of the logistic regression
print(result.summary())
```
