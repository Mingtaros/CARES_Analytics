# Encode Categorical columns to Numerical

## Age Groups
Ordinal data, do simple label encoding

```python
cares["AGEcategory"].value_counts().sort_index()
```

```python
age_group_mapping = {
    "18-29": 0,
    "30-49": 1,
    "50-64": 2,
    "65-74": 3,
    "75-84": 4,
    ">=85": 5
}
```

```python
cares.replace({"AGEcategory": age_group_mapping}, inplace=True)
cares.sample(5)
```

## Gender

```python
gender_mapping = {
    "FEMALE": 0,
    "MALE": 1
}
```

```python
cares.replace({"GENDER": gender_mapping}, inplace=True)
cares.sample(5)
```

## Race

```python
cares["RaceCategory"]
```

```python
race_category_onehot = pd.get_dummies(cares["RaceCategory"], dtype=int).drop("Others", axis=1)

cares = pd.concat([cares, race_category_onehot], axis=1)
```

```python
cares.drop("RaceCategory", axis=1, inplace=True)
```
